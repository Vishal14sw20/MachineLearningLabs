```python
plt.subplots(figsize=(10,10))
plt.subplot(2,2,1)
sns.countplot(x="International plan",hue="Churn",data=data)
plt.subplot(2,2,2)
sns.countplot(x="Voice mail plan",hue="Churn",data=data)
plt.subplot(2,2,3)
sns.countplot(x="Customer service calls",hue="Churn",data=data)
```
